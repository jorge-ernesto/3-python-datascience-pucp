## Histogram

- Un **histograma** es un gráfico univariado que nos ayuda a comprender la distribución de una variable numérica continua.
- Divide el rango de las variables continuas en intervalos de igual longitud y luego cuenta el número de observaciones en cada intervalo.
- Utilizaremos la función *histplot()* de seaborn para crear histogramas.

````python
sns.histplot(data=df, x='price')
````

**Veamos cómo podemos personalizar un histograma.**

````python
plt.title('Histogram:Price')
plt.xlim(3000,50000)
plt.ylim(0,70)
plt.xlabel('Price of cars')
plt.ylabel('Frequency')
sns.histplot(data=df, x='price',color='orange');
````

**Podemos especificar el número de intervalos (o grupos o contenedores) que deseamos crear configurando el parámetro *bins*.**

- Si no se especifica, se pasa a [numpy.histogram_bin_edges()](https://numpy.org/doc/stable/reference/generated/numpy.histogram_bin_edges.html#numpy.histogram_bin_edges).

````python
sns.histplot(data=df, x='price', bins=5)
````

````python
sns.histplot(data=df, x='price', bins=20)
````

**Si deseamos especificar el ancho de los intervalos (o grupos o contenedores), podemos utilizar el parámetro *binwidth*.**

````python
sns.histplot(data=df, x='price', binwidth=20)
````

````python
sns.histplot(data=df, x='price', binwidth=200)
````

**Además de las barras, también podemos agregar una estimación de densidad configurando el parámetro *kde* en *True*.**

- **Kernel Density Estimation**, o **KDE**, visualiza la distribución de los datos en un intervalo continuo.
- La escala convencional para el KDE es: **Frecuencia total de cada contenedor × Probabilidad**.

````python
sns.histplot(data=df, x='price', kde=True);
````

````python
sns.histplot(data=df, x='price', bins=700, kde=True);
````

Claramente, si aumentamos el número de contenedores, se reduce la frecuencia en cada grupo (contenedor). Dado que la escala de KDE depende de la frecuencia total de cada contenedor, el código anterior nos proporciona un gráfico de KDE aplanado.

**Veamos los histogramas de algunos atributos más en los datos.**

````python
sns.histplot(data=df, x='curb_weight', kde=True);
````

- Un histograma se dice que es **simétrico** si los lados izquierdo y derecho se asemejan a imágenes de espejo entre sí cuando el histograma se corta por la mitad.

````python
sns.histplot(data=df, x='horsepower', kde=True);
````

- Los grupos de barras más altos, es decir, los picos, en un histograma representan los **modos** de los datos.
- Un histograma **inclinado hacia la derecha** tiene un gran número de ocurrencias en el lado izquierdo del gráfico y pocas en el lado derecho del gráfico.
- De manera similar, un histograma **inclinado hacia la izquierda** tiene un gran número de ocurrencias en el lado derecho del gráfico y pocas en el lado izquierdo del gráfico.

**Los histogramas son intuitivos, pero rara vez son una buena elección cuando queremos comparar las distribuciones de varios grupos. Por ejemplo,**

````python
sns.histplot(data=df, x='price', hue='body_style', kde=True);
````

- También se le llama **diagrama de resumen de cinco números**, donde el resumen de cinco números incluye el valor mínimo, el primer cuartil, la mediana, el tercer cuartil y el valor máximo.
- La función *boxplot()* de seaborn se puede utilizar para crear un diagrama de caja.

image.png

## Lineplot

Supongamos que tu conjunto de datos tiene múltiples valores de y para cada valor de x. Un gráfico de líneas es una excelente manera de visualizar esto. Este tipo de datos a menudo aparece cuando tenemos datos que evolucionan con el tiempo, por ejemplo, cuando tenemos datos mensuales a lo largo de varios años. Si deseamos comparar los meses individuales, un gráfico de líneas es una excelente opción. A esto a veces se le llama análisis de estacionalidad.



image.png

- Un **gráfico de líneas** utiliza líneas rectas para conectar puntos de datos individuales y mostrar una tendencia o patrón en los datos.
    - Por ejemplo, efectos estacionales y cambios significativos a lo largo del tiempo.

- La función *lineplot()* de seaborn, de forma predeterminada, realiza una agregación de múltiples valores de y en cada valor de x y utiliza una estimación de la tendencia central para el gráfico.

- *lineplot()* asume que la mayoría de las veces estás tratando de representar y como una función de x. Por lo tanto, de forma predeterminada, ordena los datos por los valores de x antes de graficarlos.

````python
# Cargando uno de los conjuntos de datos de ejemplo disponibles en seaborn
flights = sns.load_dataset("flights")

# Creamis un lineplot
sns.lineplot(data = flights , x = 'month' , y = 'passengers');
````

- El área sombreada de color azul claro es en realidad el '**intervalo de confianza**' de las estimaciones de los valores de y para cada valor del eje x.

- El **intervalo de confianza** es un rango de valores alrededor de esa estimación que se cree que contiene el valor real de esa estimación con cierta probabilidad.

**Podemos desactivar los intervalos de confianza configurando el parámetro *ci* en *'False'*.**

````python
sns.lineplot(data = flights , x = 'month' , y = 'passengers', errorbar=('ci',False));
````

**También podemos verificar la relación entre dos variables para diferentes categorías especificando el parámetro *hue*.**

````python
sns.lineplot(data=flights,x = 'month' , y = 'passengers',errorbar=('ci', False) ,hue='year');
````

**Podemos cambiar el estilo de las líneas agregando el parámetro 'style' a la función.**

````python
# Cargando uno de los conjuntos de datos de ejemplo disponibles en seaborn
fmri = sns.load_dataset("fmri")

sns.lineplot(data = fmri, x="timepoint", y="signal", hue="region", style="region", errorbar=('ci', False));
````

**También podemos agregar marcadores en cada observación para identificar grupos de una manera más clara.**

````python
sns.lineplot(data = fmri, x="timepoint", y="signal", hue="region", style="region", errorbar=('ci', False), markers = True);
````

**Personalicemos el gráfico de líneas para una mejor visualización.**

````python
plt.figure(figsize = (15,7))
sns.lineplot(data = flights , x = 'month' , y = 'passengers', hue = 'year')
plt.ylabel('Number of Passengers')
plt.legend(bbox_to_anchor=[1, 1]); # Otra forma de cambiar la ubicación de la leyenda en el gráfico
````

- Ten en cuenta que, a diferencia de los gráficos de barras e histogramas, los gráficos de líneas pueden no incluir una línea de base en cero.
- Creamos un gráfico de líneas para enfatizar los cambios en el valor, en lugar de la magnitud de los propios valores, por lo que una línea de base en cero no es significativa.

## Scatterplot

A veces queremos saber si dos variables tienen algún significado cuando se ponen juntas, si un pequeño cambio en una variable afecta a la otra variable. En tales casos, trazar un **gráfico de dispersión**, o **diagrama de dispersión**, con nuestros puntos de datos puede ayudarnos a verificar si existe una posible relación entre ellas.

- Un **gráfico de dispersión** es la forma más simple de representación diagramática de dos variables.
- Toma dos ejes perpendiculares de coordenadas, uno para x y otro para y.
- A diferencia del gráfico de líneas, traza directamente cada par de valores como un punto en el espacio 2D.
- La función *scatterplot()* de seaborn se puede utilizar para crear un gráfico de dispersión.

````python
sns.scatterplot(data=df, x='engine_size', y='horsepower');
````

**También podemos verificar la relación entre dos variables para diferentes categorías especificando el parámetro *hue*.**

````python
sns.scatterplot(data=df, x='engine_size', y='horsepower', hue='fuel_type');
````

**Podemos asignar la misma variable como *hue* a otro parámetro *style* que variará los marcadores y creará un gráfico más legible.**

````python
sns.scatterplot(data=df, x='engine_size', y='horsepower', hue='fuel_type', style='fuel_type');
````

**Correlation**

La correlación significa asociación. Más precisamente, expresa en qué medida dos variables cambian juntas a una tasa constante.

- En un gráfico de dispersión, cuando la variable y tiende a aumentar a medida que la variable x aumenta, decimos que hay una **correlación positiva** entre las variables.
- De nuevo, cuando la variable y tiende a disminuir a medida que la variable x aumenta, decimos que hay una **correlación negativa** entre las variables.
- Si los puntos en el gráfico de dispersión parecen estar dispersos al azar, decimos que no hay **correlación** entre las variables.

**Veamos la relación entre algunas variables adicionales utilizando gráficos de dispersión.**

````python
sns.scatterplot(data=df, x='curb_weight', y='engine_size');
````

A partir del gráfico anterior, podemos decir que estas variables están *correlacionadas positivamente*.

````python
sns.scatterplot(data=df, x='bore', y='stroke');
````

- A partir del gráfico anterior, es evidente que las variables no tienen correlación.

**Nota:**
1. Una correlación fuerte tendrá puntos de datos cercanos entre sí, mientras que una correlación débil tendrá puntos de datos más alejados.
2. No podemos medir la relación cuantitativamente utilizando un gráfico de dispersión. Solo proporciona una expresión del cambio relativo entre las variables.

Podemos ver a partir del gráfico de dispersión de `engine_size` vs `horsepower` que hay una correlación positiva entre las dos variables. Ahora, queremos medir la relación entre estas dos variables de manera cuantitativa y tratar de predecir '**horsepower**' en función de '**engine size**'. Esto se puede hacer fácilmente ajustando un modelo lineal. Aquí entra en juego la función ***lmplot()*** de seaborn para ayudarnos con eso.

## lmplot

Hemos visto que los histogramas nos ayudan a comprender la distribución de variables individuales y los gráficos de dispersión nos ayudan a identificar la relación entre dos variables. Aunque podemos ver y analizar estos gráficos por separado, tenerlos juntos en una sola visualización nos permitiría capturar mucha más información de manera concisa. Es aquí donde un **jointplot** viene en nuestra ayuda.

## Joint plot

* Un jointplot ayuda a visualizar rápidamente los perfiles bivariados y univariados en el mismo gráfico.
* Consta de tres gráficos: uno muestra la relación entre dos variables, mientras que los otros dos muestran la distribución individual de cada variable en los márgenes.
* La función *jointplot()* de seaborn se puede utilizar para crear un jointplot.

````python
sns.jointplot(data=df, x='engine_size', y='horsepower');
````

- De forma predeterminada, *jointplot()* crea un gráfico de dispersión para mostrar la relación entre dos variables y se crean histogramas individuales en los márgenes de los ejes x e y para cada una de las variables.

Una de las desventajas de los **gráficos de dispersión** es la superposición de puntos. Cuando tenemos grandes volúmenes de datos para representar, los puntos de datos en el gráfico se superponen entre sí, lo que dificulta la interpretación de los datos.

En tales casos, podemos *dividir todo el gráfico en contenedores utilizando diferentes formas (cuadrados, triángulos, hexágonos, etc.)* y luego intentar ver el número de puntos de datos que caen en cada uno de los contenedores.

La función *jointplot()* de seaborn proporciona un tipo '**hex**' para representar los datos de la manera mencionada anteriormente.

**Podemos crear un gráfico de hexágonos configurando *kind="hex"*.**

- En realidad, divide la ventana de trazado en varios hexágonos y el color de cada hexágono denota el número de puntos de datos en él.

````python
sns.jointplot(data=df, x='engine_size', y='horsepower', kind="hex");
plt.colorbar(); # añade un eje separado que indica la escala de colores en este gráfico
````

**También podemos crear un gráfico KDE configurando *kind="kde"*.**

- Un gráfico KDE jointplot también utiliza el color para determinar dónde las observaciones son más densas.
- Crea un gráfico continuo utilizando probabilidades de cómo debería introducirse nuevos datos.

````python
sns.jointplot(data=df, x='engine_size', y='horsepower', kind="kde", fill=True);
````

**Veamos los gráficos conjuntos para algunos atributos adicionales en los datos.**
- Utilizando **kind = "reg"**, podemos utilizar el gráfico conjunto para encontrar la mejor línea o curva que se ajuste al gráfico.

````python
sns.jointplot(data=df, x='price', y='city_mpg', kind="reg");
````

## Violin Plot

* Un **gráfico de violín** nos permite visualizar la distribución de una variable numérica para varios grupos.
* Es similar a un **diagrama de caja**, pero con un gráfico girado en cada lado, lo que proporciona más información sobre la estimación de densidad en el eje y.
* La función *violinplot()* de seaborn se puede utilizar para crear un gráfico de violín.

````python
sns.violinplot(data=df, x='horsepower');
````

- El punto blanco en el gráfico representa la **mediana**.
- La barra gris gruesa en el centro representa el **rango intercuartil**.
- La sección más ancha del gráfico de violín representa una probabilidad más alta y las secciones más delgadas representan una probabilidad más baja para el valor dado.

**Podemos obtener un gráfico vertical configurando el parámetro *orient* en *'v'* y asignando una variable numérica al eje y.**

````python
sns.violinplot(data=df, x='fuel_type', y='horsepower', orient='v');
````

**Veamos los gráficos de violín para algunos atributos adicionales en los datos.**

````python
sns.violinplot(data=df, x='engine_size', y='fuel_type');
````

**Podemos usar el parámetro *palette* para cambiar la paleta de colores.**

- seaborn tiene seis variaciones de ella: deep, muted, pastel, bright, dark, colorblind.

````python
sns.violinplot(data=df, x='body_style', y='engine_size', palette="bright");
````

````python
sns.violinplot(data=df, x='engine_location', y='price', palette="colorblind");
````

Los histogramas y otros gráficos de distribución suelen ser preferidos para conjuntos de datos más grandes. Cuando tenemos un conjunto de datos pequeño y queremos visualizar la distribución de frecuencia, se puede utilizar un **gráfico de puntos**.

## Strip Plot

* Un gráfico de puntos es básicamente un gráfico de dispersión que diferencia diferentes categorías.
* La función *stripplot()* de seaborn se puede utilizar para crear un gráfico de puntos.

````python
sns.stripplot(data=df, x='engine_size');
````

Los gráficos de puntos se consideran una buena alternativa a un diagrama de caja o un gráfico de violín para comparar las distribuciones de datos cuando tenemos menos puntos de datos.

````python
plt.figure(figsize=(12,5))
sns.stripplot(data=df, x='body_style', y='engine_size');
````

**Los gráficos de puntos son más útiles cuando agregamos ruido aleatorio llamado "jitter" para evitar la superposición de puntos de datos con los mismos valores.**

````python
plt.figure(figsize=(16,5))
sns.stripplot(data=df, x='body_style', y='engine_size', jitter=True);
````

**Podemos configurar el parámetro *hue* para mostrar observaciones de subgrupos.**

````python
plt.figure(figsize=(10,7))
sns.stripplot(data=df, x='fuel_type', y='engine_size', hue="number_of_doors", jitter=True);
````

**Veamos los gráficos de puntos para algunos atributos adicionales en los datos.**

````python
plt.figure(figsize=(10,7))
sns.stripplot(data=df, x='number_of_doors', y='price',jitter=True);
````

````python
plt.figure(figsize=(10,7))
sns.stripplot(data=df, x='number_of_doors', y='horsepower', jitter=True);
````

En un gráfico de puntos, los puntos se distribuyen de manera aleatoria en la dirección vertical, lo que significa que no podemos utilizar el ancho para estimar la distribución. Esto se puede hacer fácilmente con los **gráficos de dispersión (swarm plots)**.

## Swarm Plot

* Un **gráfico de dispersión (swarm plot)** es muy similar al gráfico de puntos, pero evita la superposición de puntos.
* La función *swarmplot()* de seaborn se puede utilizar para crear dicho gráfico para valores categóricos.

````python
sns.swarmplot(data=df, x='number_of_doors', y='price');
````

- En un gráfico de dispersión, los puntos se disponen en columnas verticales.
- Por lo tanto, podemos estimar la distribución de nuestros datos observando el ancho del gráfico.

**Podemos agregar el parámetro *hue* a *swarmplot()* y separar las categorías.**

````python
sns.swarmplot(data=df, x='fuel_type', y='price', hue='number_of_doors');
````

**Podemos separar los dos productos configurando el argumento *dodge* en *True*.**

````python
sns.swarmplot(data=df, x='fuel_type', y='price', hue='number_of_doors', dodge=True);
````

## Catplot

* La función *catplot()* proporciona un nuevo marco que brinda acceso a varios tipos de gráficos que muestran la relación entre una variable numérica y una o más variables categóricas.
* Catplot admite 8 gráficos diferentes, incluidos *stripplot()*, *swarmplot()*, *boxplot()* y más, con el tipo predeterminado siendo 'strip'.

````python
sns.catplot(data=df, x='fuel_type', y='horsepower');
````

**Podemos establecer *kind='point'* para crear un pointplot utilizando la función *catplot()**.

## Pair Plot

* Un **pairplot** muestra la relación entre dos variables numéricas para cada par de columnas en el conjunto de datos.
* Crea una cuadrícula de ejes de manera que cada variable en los datos se compartirá en el eje y a lo largo de una fila única y en el eje x a lo largo de una columna única.
* La función *pairplot()* de seaborn se puede utilizar para crear dicho gráfico.

````python
sns.pairplot(data=df[['normalized_losses','wheel_base','curb_weight','engine_size','price','peak_rpm']])
````

**Podemos agregar el parámetro *hue* en pairplot para crear un mapeo semántico.**

- Cambia el gráfico marginal predeterminado a un gráfico de densidad (kde) superpuesto.

**También podemos agregar el parámetro *vars* para asignar una lista de variables del conjunto de datos para las cuales queremos crear el pairplot.**

````python
sns.pairplot(data=df, vars=['wheel_base', 'curb_weight', 'engine_size', 'price'], hue='number_of_doors');
````

**Podemos configurar *corner=True* para trazar solo el triángulo inferior de un pairplot.**

````python
sns.pairplot(data=df, vars=['wheel_base', 'curb_weight', 'engine_size', 'price'], corner=True);
````

## Heatmap

* Un **mapa de calor (heatmap)** es una representación gráfica de los datos en forma de matriz codificada por colores.
* Es una excelente manera de representar la correlación entre cada par de columnas en los datos.
* La función *heatmap()* de seaborn nos ayuda a crear este tipo de gráfico.

````python
sns.heatmap(data=df[['wheel_base','curb_weight','engine_size','price']].corr());
````

**Podemos configurar el parámetro *annot* en *True* para mostrar el valor numérico en cada celda.**

- Para quitar la barra de colores, se puede establecer el parámetro *cbar* en *False*.

````python
sns.heatmap(data=df[['wheel_base','curb_weight','engine_size','price']].corr(), annot=True, cbar=False);
````

**Podemos aplicar un mapa de colores diferente con el parámetro *cmap* para obtener un mejor atractivo visual.**

````python
sns.heatmap(data=df[['wheel_base','curb_weight','engine_size','price']].corr(), annot=True, cmap='YlGnBu');
````

## Personalización de gráficos

- En esta sección veremos cómo podemos personalizar los gráficos utilizando diferentes funciones de matplotlib y seaborn.

Para cambiar las escalas de los ejes de un gráfico:

```
    plt.xlim(left, right)   # here left and right indicates the min and the max limits of  x-axis respectively
    plt.ylim(bottom, top)   # here bottom and top indicates the min and the max limits of  y-axis respectively
```

Para cambiar el título del gráfico y las etiquetas de los ejes:
```
    plt.title()
    plt.xlabel()   # for x_axis labels
    plt.ylabel()   # for y_axis labels
```

Para mostrar las líneas de la cuadrícula en un gráfico:

```
    sns.set(style="darkgrid")
    sns.set(style="whitegrid")
    sns.set(style="white")
```

Para ocultar los ejes de un gráfico:

```
    plt.axis('off')   # it will hide both the x-axis and y-axis
```

Para personalizar las etiquetas de la leyenda del gráfico:

```
    ax.legend([])   # takes list of the legend values
```

Para cambiar la posición de la leyenda del gráfico:

```
    plt.legend(loc=___)
```

Se pueden agregar los siguientes valores a *loc* para cambiar la ubicación de la leyenda:

- 'best'
- 'upper right'
- 'upper left'
- 'lower left'
- 'lower right'
- 'right'
- 'center left'
- 'center right'
- 'lower center'
- 'upper center'
- 'center'
